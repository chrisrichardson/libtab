```python
import numpy
import matplotlib.pyplot as plt
from libtab import Lagrange, CellType, create_lattice

L = Lagrange(CellType.triangle, 2)

pts = create_lattice(CellType.triangle, 50, True)
w0 = L.tabulate(2, pts)
```

```python
w = w0[0]

fig = plt.figure(figsize=(12, 8))
ax = fig.subplots(2, 3)
ax[0, 0].set_title("Shape functions N(x,y)")
for j, a in enumerate(ax.flatten()):
    shape_fn = w[:, j]
    a.tricontourf(pts[:, 0], pts[:, 1], shape_fn)
plt.show()
```

```python
w = w0[1]

fig = plt.figure(figsize=(12, 8))
ax = fig.subplots(2, 3)
ax[0, 0].set_title("Shape function derivative: $dN/dx$")
for j, a in enumerate(ax.flatten()):
    shape_fn = w[:, j]
    a.tricontourf(pts[:, 0], pts[:, 1], shape_fn, levels=numpy.arange(-5.0, 5.0, 0.8))
plt.show()
```

```python
w = w0[2]

fig = plt.figure(figsize=(12, 8))
ax = fig.subplots(2, 3)
ax[0, 0].set_title("Shape function derivative: $dN/dy$")
for j, a in enumerate(ax.flatten()):
    shape_fn = w[:, j]
    a.tricontourf(pts[:, 0], pts[:, 1], shape_fn, levels=numpy.arange(-5.0, 5.0, 0.8))
plt.show()
```
